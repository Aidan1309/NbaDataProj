```python
def find_ap(combo): #for finding error metric
    actual = combo.sort_values("Share", ascending=False).head(5) #sort the data by highest mvp vote share and take the top 5 players
    predicted = combo.sort_values("predictions", ascending=False) #look at predictions and sort them
    ps = []
    found = 0
    seen = 1
    for index, row in predicted.iterrows():
        if row["Player"] in actual["Player"].values: #if the player in the predicted top 5 is in the actual top 5
            found += 1
            ps.append(found/seen)
        seen += 1
    return sum(ps) / len(ps)
```

```python
years = list(range(1995,2023))
```

```python
aps= []
all_predictions = []
for year in years[5:]: # looking from years 20001 and on, run the same code to find the error metric as before
    train = stats[stats["Year"] <year]
    test = stats[stats["Year"] == year]
    reg.fit(train[predictors], train["Share"])
    predictions = reg.predict(test[predictors])
    predictions = pd.DataFrame(predictions, columns=["predictions"], index=test.index)
    combo = pd.concat([test[["Player", "Share"]], predictions], axis=1)
    all_predictions.append(combo)
    aps.append(find_ap(combo))
```

```python
sum(aps) / len(aps)
```

```python
def add_ranks(combo): #add in diagnostics for the machine learning algorithm
    combo = combo.sort_values("Share", ascending=False)
    combo["Rk"] = list(range(1,combo.shape[0]+1))
    combo = combo.sort_values("predictions", ascending=False)
    combo["Predicted_Rk"] = list(range(1,combo.shape[0]+1))
    combo["Diff"] = combo["Rk"] - combo["Predicted_Rk"]
    return combo
```

```python
ranking = add_ranks(all_predictions[1])
ranking[ranking["Rk"] < 6].sort_values("Diff", ascending=False)
```

```python
def backtest(stats, model, year, predictors): #single function for backtesting
    aps= []
    all_predictions = []
    for year in years[5:]: # looking from years 20001 and on, run the same code to find the error metric as before
        train = stats[stats["Year"] <year]
        test = stats[stats["Year"] == year]
        model.fit(train[predictors], train["Share"])
        predictions = reg.predict(test[predictors])
        predictions = pd.DataFrame(predictions, columns=["predictions"], index=test.index)
        combo = pd.concat([test[["Player", "Share"]], predictions], axis=1)
        combo = add_ranks(combo)
        all_predictions.append(combo)
        aps.append(find_ap(combo))
    return sum(aps)/len(aps), aps, pd.concat(all_predictions) #mean average percision (sum(aps)/len(aps))
```

```python
mean_ap, aps, all_predictions = backtest(stats, reg, years[5:], predictors)
```
